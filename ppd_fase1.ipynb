```python
import csv
```

Open file and ensuring encoding to UTF-8 for no Unicode errors.
Printing a sample of size 5.

---



```python
with open('steam_games.csv', newline='', encoding='utf-8') as f:
    reader = csv.reader(f)
    my_dict = {}

    for i in reader:
      try:
        my_dict[i[2][-4:]] += 1
      except KeyError:
        my_dict[i[2][-4:]] = 1

      try:
        if i[6] == "0.0":
          my_dict['free'] += 1
        else:
          my_dict['paid'] += 1
      except KeyError:
        if i[6] == "0.0":
          my_dict['free'] = 1
        else:
          my_dict['paid'] = 0

      try:
        my_dict["count"] += 1
      except KeyError:
        my_dict["count"] = -1
```

```python
more_realeses = 0
year_games = []

for k, v in my_dict.items():
  if v >= more_realeses:
    year_games = []
    more_realeses = v
    year_games.append((k, v))
print(more_realeses)
print(year_games)
print(my_dict)
```
