# **REQUIREMENTS**

### Required imports.

```python
import csv
```

###Open file and ensuring encoding to UTF-8 for no Unicode errors.<br>Create dictionary holding all information needed for the project.<br>Dictonary also increase by far the performance of all queries.



```python
with open('steam_games.csv', newline='', encoding='utf-8') as f:
    reader = csv.reader(f)
    #Criando dicionário com as colunas a serem usadas para otimização
    my_dict = {'years': {}}

    for i in reader:
      #adiciona 1 a key years[ano] para toda vez que o ano aparece no dataset
      try:
        my_dict['years'][i[2][-4:]] += 1
      except KeyError:
        my_dict['years'][i[2][-4:]] = 1

      #jogos gratis/pagos, free_palyers/paid_palyer, total_num_players
      #except else comeca com 0 e nao 1 por que o primeiro row do dataset sao os headers das colunas e nao valores
      try:
        if i[6] == "0.0":
          my_dict['free'] += 1
          my_dict['free_players'] += int(i[4])
          my_dict['player_count'] += int(i[4])
        else:
          my_dict['paid'] += 1
          my_dict['paid_players'] += int(i[4])
          my_dict['player_count'] += int(i[4])
      except KeyError:
        if i[6] == "0.0":
          my_dict['free'] = 1
          my_dict['free_players'] = int(i[4])
          my_dict['player_count'] = int(i[4])
        else:
          my_dict['paid'] = 0
          my_dict['paid_players'] = 0
          my_dict['player_count'] = 0

      # total de jogos no dataset
      #except comeca com 0 e nao 1 por que o primeiro row do dataset sao os headers das colunas e nao valores
      try:
        my_dict["count"] += 1
      except KeyError:
        my_dict["count"] = 0
```

```python
more_games = 0 #more_games checa qual key em years tem o valor mais alto
year_games = [] #salva um set com o a key years (ano) e seu valor (more_games)

for k, v in my_dict['years'].items():
  if v >= more_games:
    year_games = []
    more_games = v
    year_games.append((k, v))
```

```python
#PERGUNTA 1
print(f"Total de jogos = {my_dict["count"]} [100%]\n"
      f"Total de jogos pagos = {my_dict['paid']} [{my_dict['paid'] * 100 / my_dict['count']:.2f}%]\n"
      f"Total de jogos grátis = {my_dict['free']} [{my_dict['free'] * 100 / my_dict['count']:.2f}%]")

#PERGUNTA 2
print(f"\n{year_games[0][0]} foi o ano com mais lançamentos com {year_games[0][1]} jogos!")

#PERGUNTA 3
print(f"\nDentro deste dataset o número máximo de usuários simultâneos que estavam jogando todos os jogos é {my_dict["player_count"]}.\n"
      f"De todos esses jogadores:\n\n"
      f"    {my_dict["paid_players"]} [{my_dict['paid_players'] * 100 / my_dict['player_count']:.2f}%] jogam em jogos pagos\n"
      f"    {my_dict["free_players"]} [{my_dict['free_players'] * 100 / my_dict['player_count']:.2f}%] jogam em jogos grátis")
```
